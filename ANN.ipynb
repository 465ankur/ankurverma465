```python
# Import necessary libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load dataset
file_path = r"D:\Data since couress\Neural networks\Neural networks\Alphabets_data.csv"
df = pd.read_csv(file_path)

# Explore dataset
print("Dataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

# Handle missing values (if any)
df.dropna(inplace=True)  # Removing rows with missing values

# Separate features and labels
X = df.drop(columns=["letter"])  # Exclude categorical column
y = df["letter"]  # Target variable (categorical)

# Encode categorical labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Convert categorical labels to numeric

# Normalize feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Build an improved ANN model with the corrected input layer
model = keras.Sequential([
    Input(shape=(X_train.shape[1],)),  # ✅ Corrected input layer
    Dense(128, activation="relu"),
    Dense(64, activation="relu"),
    Dense(len(label_encoder.classes_), activation="softmax")  # Multi-class classification
])

# Compile model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# Train model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Make predictions
y_pred = np.argmax(model.predict(X_test), axis=1)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average="weighted")
recall = recall_score(y_test, y_pred, average="weighted")
f1 = f1_score(y_test, y_pred, average="weighted")

# Display evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```
