```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset correctly
df = pd.read_excel(r"D:\Data since couress\Random Forest\Random Forest\glass.xlsx", skiprows=1)  # Adjust skiprows if headers are duplicated

# Strip column names to remove any extra spaces
df.columns = df.columns.str.strip()

# Convert all columns to numeric, handling errors gracefully
df = df.apply(pd.to_numeric, errors='coerce')

# Drop rows containing NaN values introduced during conversion
df.dropna(inplace=True)

# Check data types to confirm corrections
print(df.dtypes)

# Select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Plot histograms for numerical features
df[numeric_cols].hist(figsize=(12, 10))
plt.show()

# Box plot to check outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numeric_cols])
plt.xticks(rotation=45)
plt.show()
```
